```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("data.csv", delimiter=';')
df = dataset.tail(10)
df
```

```python
df = df.iloc[:,:].values
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

country = LabelEncoder()

df[:,8] = country.fit_transform(df[:,8])

df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ct = ColumnTransformer(transformers=[('encode', ohe, [8])], remainder='passthrough')
```

```python
y = ct.fit_transform(df)
```

```python
y
```

```python
ds = pd.DataFrame(y)
ds
```
